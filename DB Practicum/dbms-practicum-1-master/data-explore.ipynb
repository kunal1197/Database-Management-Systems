```python
import pandas as pd
```

```python
df = pd.read_csv("BirdStrikesData.csv")
```

```python
col_names = df.columns.tolist()
```

```python
type_and_val = {}

for col_name in col_names:
    headers = col_name.split(":")
    if(len(headers) == 1):
        type_and_val[headers[0]] = [headers[0]]
    else:
        if(type_and_val.get(headers[0])):
            type_and_val[headers[0]].append(headers[1])
        else:
            type_and_val[headers[0]] = [headers[1]]
```

```python
origin_states = []
for i, row in df.iterrows():
    if type(row['Airport: Name']) == float:
        print(row['Record ID'], row['Origin State'])
        origin_states.append(row['Origin State'])
        
print(len(origin_states), set(origin_states))
```

```python
states_and_airports = {}
for i, row in df.iterrows():
    if row['Origin State'] in states_and_airports.keys():
        states_and_airports[row['Origin State']].append(row['Airport: Name'])
    else:
        states_and_airports[row['Origin State']] = [row['Airport: Name']]
        
states_and_airports = {state: list(set(airport)) for state, airport in states_and_airports.items()}
```

```python
states_and_airports['Louisiana']
```

```python
vals_list = []
for i, row in df.iterrows():
    if type(row['Altitude bin']) == float:
        vals_list.append(type(row['Is Aircraft Large?']) == float)
        
print(len(set(vals_list)))
```

```python
make_model_and_engines = {}
for i, row in temp_df.iterrows():
    if row['Aircraft: Make/Model'] in make_model_and_engines.keys():
        make_model_and_engines[row['Aircraft: Make/Model']].append(row['Aircraft: Number of engines?'])
    else:
        make_model_and_engines[row['Aircraft: Make/Model']] = [row['Aircraft: Number of engines?']]
        
make_model_and_engines = {make_model: list(set(engines)) for make_model, engines in make_model_and_engines.items()}
```

```python
make_model_and_engines
```

```python
for i, row in temp_df.iterrows():
    if type(row['Aircraft: Number of engines?']) == float:
        print(row['Aircraft: Make/Model'], "-", make_model_and_engines[row['Aircraft: Make/Model']])
```

```python
airport_and_origin_state = {}
for i, row in temp_df.iterrows():
    if row['Airport: Name'] in airport_and_origin_state.keys():
        airport_and_origin_state[row['Airport: Name']].append(row['Origin State'])
    else:
        airport_and_origin_state[row['Airport: Name']] = [row['Origin State']]
        
airport_and_origin_state = {airport: list(set(origin_state)) for airport, origin_state in airport_and_origin_state.items()}
```

```python
for i, row in temp_df.iterrows():
    if type(row['Origin State']) == float:
        print(row['Origin State'], "->", row['Airport: Name'], "->", airport_and_origin_state[row['Airport: Name']])
```

```python
for airport, origin_state in airport_and_origin_state.items():
    if(type(origin_state[0]) == float):
        print(airport)
```
